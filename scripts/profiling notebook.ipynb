```python
import pandas as pd
from ydata_profiling import ProfileReport
import os
```

```python
for folder in [r'data/Project Dataset-20241024T131910Z-001']:
    for filename in os.listdir(folder):
        if filename.endswith('.csv'):
            file_path = os.path.join(folder, filename)
            df = pd.read_csv(file_path)
            profile = ProfileReport(df, title=f'Profiling Report for {filename}', explorative=True)
            report_path = os.path.join('reports', f'{filename}_profiling_report.html')
            profile.to_file(report_path)
        
        elif filename.endswith('.parquet'):
            file_path = os.path.join(folder, filename)
            df = pd.read_csv(file_path)
            profile = ProfileReport(df, title=f'Profiling Report for {filename}', explorative=True)
            report_path = os.path.join('reports', f'{filename}_profiling_report.html')
            profile.to_file(report_path)

        elif filename.endswith('.pickle'):
            file_path = os.path.join(folder, filename)
            df = pd.read_csv(file_path)
            profile = ProfileReport(df, title=f'Profiling Report for {filename}', explorative=True)
            report_path = os.path.join('reports', f'{filename}_profiling_report.html')
            profile.to_file(report_path)
        
        elif filename.endswith('.json'):
            file_path = os.path.join(folder, filename)
            df = pd.read_csv(file_path)
            profile = ProfileReport(df, title=f'Profiling Report for {filename}', explorative=True)
            report_path = os.path.join('reports', f'{filename}_profiling_report.html')
            profile.to_file(report_path)
```
